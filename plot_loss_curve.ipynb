```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
csv_file = "epoch-wise_Training_Losses_Data.csv"
data = pd.read_csv(csv_file)

# Extract columns
epochs = data['epoch']
total_loss = data['loss']

# Plot the data with a logarithmic scale
plt.figure(figsize=(10, 6))
plt.plot(epochs, total_loss, marker='o', label='Total Loss', linestyle='-', linewidth=2)
# plt.yscale('log') 
plt.xlabel('Epoch')
plt.ylabel('Total Loss (avg batch wise)')
plt.title('Epoch vs. Total Loss (avg batch wise)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()

# Save the plot as an image
plot_filename = "epoch_vs_total_loss_log_scale.png"
plt.savefig(plot_filename, dpi=300)

# Show the plot
plt.show()

print(f"Plot saved as {plot_filename}")
```
